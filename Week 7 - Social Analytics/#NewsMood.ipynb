```python
#define list color    
color_list = []
for index, row in sentiments_pd.iterrows():
    if row["Source"] == "BBC":
        color = "lightskyblue"
        color_list.append(color)
    elif row["Source"] == "CNN":
        color ="green"
        color_list.append(color)
    elif row["Source"] == "FoxNews":
        color = "red"
        color_list.append(color)
    elif row["Source"] == "CBSNews":
        color ="blue"
        color_list.append(color)
    elif row["Source"] == "nytimes":
        color = "yellow"
        color_list.append(color)
        
```

```python
# Convert color_list to DataFrame
color_list_pd = pd.DataFrame({"Colors":color_list})
sentiments_pd["Colors"] = color_list_pd["Colors"]

sentiments_pd.head(150)
```
